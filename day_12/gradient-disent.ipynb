df/dx=2*x

```python
import numpy as np
import matplotlib.pyplot as plt
def plot_graph(x,y,x_label,y_label,colour):
    fig, ax = plt.subplots()
    plt.ylabel(y_label)
    plt.xlabel(x_label)
    ax.plot(x, y,colour)
    plt.show()
```

```python
lr=0.1
x=11
X=[]
Y=[]
for i in range(50): #Iteration number is 50
    X.append(x) #store the result value in X
    y=x**2
    Y.append(y)
    #Display the optimized result
    x = x - lr*2*x #find the optimal result using gradient descent

#Create the data between -11 and +11 to draw the x^2 line
x = np.linspace(-11, 11, 100)
y = x**2
plot_graph(x,y,"X","X^2",'-b')

fig, ax = plt.subplots()
plt.ylabel('Y Result')
plt.xlabel('Iteration')
ax.plot(Y,'r-')
plt.show()

fig, ax = plt.subplots()
plt.ylabel('X Result')
plt.xlabel('Iteration')
ax.plot(X,'r-')
plt.show()

fig, ax = plt.subplots()
ax.plot(x, y, X,'r-')
plt.show()
```
